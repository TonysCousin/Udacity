### Handling variable image sizes

Clearly, from above, we have a wide range of image sizes, by almost 10x.  Cropping is not a good answer to this, because many of the images are framed tightly around the artifact of interest, so even a little cropping will be throwing away valuable information.  Other alternatives, which might be combined:

1) Downsample the larger images - probably useful for a certain range of downsampling (e.g. up to 4x?) but we stand to lose too much info in the largest images if they are downsampled a lot.  See https://scikit-image.org/docs/dev/auto_examples/transform/plot_rescale.html

2) Throw out the smallest and/or largest images as outliers - if there are only a few way out on the tails of the distribution, this could help reduce the magnitude of the problem, but the statistics above only show that there are at least two images at each of the very extremes.

3) Build a network that handles variable image sizes - there are some techniques for this, but it involves advanced CNN construction, and adds difficulty to the final FC classification layer.  See:

### Loading image data of variable sizes

* There is an interesting answer for how to do this at https://discuss.pytorch.org/t/torchvision-and-dataloader-different-images-shapes/41026/3. Code from this page is copied below for ease of reading (without scroll bars).
* Another, maybe more useful answer, with links to lots of other answers: https://stackoverflow.com/questions/55041080/how-does-pytorch-dataloader-handle-variable-size-data

```python
##### for image exploration only
### I don't think I want to use this, as it has masks that appear to be a substitute for resizing the images,
### but it looks like it requires each image to be in a fixed size tensor, which I don't want to do.

def default_collate(batch):
    """
    Override `default_collate` https://pytorch.org/docs/stable/_modules/torch/utils/data/dataloader.html#DataLoader

    Reference:
    def default_collate(batch) at https://pytorch.org/docs/stable/_modules/torch/utils/data/dataloader.html#DataLoader
    https://discuss.pytorch.org/t/how-to-create-a-dataloader-with-variable-size-input/8278/3
    https://github.com/pytorch/pytorch/issues/1512

    We need our own collate function that wraps things up (imge, mask, label).

    In this setup,  batch is a list of tuples (the result of calling: img, mask, label = Dataset[i].
    The output of this function is four elements:
        . data: a pytorch tensor of size (batch_size, c, h, w) of float32 . Each sample is a tensor of shape (c, h_,
        w_) that represents a cropped patch from an image (or the entire image) where: c is the depth of the patches (
        since they are RGB, so c=3),  h is the height of the patch, and w_ is the its width.
        . mask: a list of pytorch tensors of size (batch_size, 1, h, w) full of 1 and 0. The mask of the ENTIRE image (no
        cropping is performed). Images does not have the same size, and the same thing goes for the masks. Therefore,
        we can't put the masks in one tensor.
        . target: a vector (pytorch tensor) of length batch_size of type torch.LongTensor containing the image-level
        labels.
    :param batch: list of tuples (img, mask, label)
    :return: 3 elements: tensor data, list of tensors of masks, tensor of labels.
    """
    data = torch.stack([item[0] for item in batch])
    mask = [item[1] for item in batch]  # each element is of size (1, h*, w*). where (h*, w*) changes from mask to another.
    target = torch.LongTensor([item[2] for item in batch])  # image labels.

    return data, mask, target
```

```python
import PIL
from PIL import Image

# Defining a custom collator for the DataLoader since its default assumes all raw images are identical
# size.  In our case they are not.  This function not only loads different size images, but manipulates
# them to be identical size (downsampling and cropping), because the CNN being used in this project
# is assuming that all images are identical size.
# This function is modified from https://jdhao.github.io/2017/10/23/pytorch-load-data-and-make-batch/
#
# `batch` is a list of tuple where first element is image tensor and
# second element is corresponding label
#
# return is a tuple of (data, target), where data is a tensor of shape [b, c, h, w] and
# target is a list of labels

AR_GOAL = 1.333333 # represents 768x1024
WIDTH_GOAL = HEIGHT_GOAL * AR_GOAL

def variable_collate(batch):
    data_list = []
    for item in batch:
        
        # get info on the raw image shape
        raw_image = item[0]
        h = raw_image.shape[1]
        w = raw_image.shape[2]
        ar = w/h
        
        # if the aspect ratio is larger than desired, then we scale by the height, otherwise by width
        if ar > AR_GOAL:
            scale_factor = HEIGHT_GOAL / h
        else:
            scale_factor = WIDTH_GOAL / w
        #print("In variable_collate: raw h = ", h, ", w = ", w, ", scale_factor = ", scale_factor)
        #print("  raw_image = ", raw_image.shape)
            
        # convert the raw_image tensor into a PIL Image object
        image = transforms.ToPILImage()(raw_image.squeeze_(0))
        #print("  After transform, image = ", image.size)
        
        # scale it, preserving aspect ratio
        resized = image.resize((int(w*scale_factor), int(h*scale_factor)))
        width = resized.width
        height = resized.height
        #print("  resized w = ", width, ", h = ", height)
            
        # now we have an image that doesn't necessarily match our goal aspect ratio, but is at least
        # as large as our goal dimensions in both directions; need to center-crop it - 
        # DON'T try to make it square!
        if width > WIDTH_GOAL:
            left = (width - WIDTH_GOAL) // 2
        else:
            left = 0
        right = left + WIDTH_GOAL #don't subtract 1; this is the next pixel beyond the edge of the image
        
        if height > HEIGHT_GOAL:
            upper = (height - HEIGHT_GOAL) // 2
        else:
            upper = 0
        lower = upper + HEIGHT_GOAL #don't subtract 1
        #print("  cropping to left = ", left, ", right = ", right, ", upper = ", upper, ", lower = ", lower)
        
        cropped_image = resized.crop((left, upper, right, lower))
        
        # convert the properly sized image back to a tensor
        cropped_tensor = transforms.ToTensor()(cropped_image).unsqueeze_(0)
        data_list.append(cropped_tensor)
        #print("  cropped_tensor = ", cropped_tensor.shape)
        

    # assemble all of the images into the data tensor for output
    data = torch.stack(data_list).squeeze()
    #print("  returning data = ", data.shape)
    
    # gather the target labels for each image
    target = [item[1] for item in batch]
    target = torch.LongTensor(target)
    
    return [data, target]
```

```python
# define dataloader parameters
batch_size = 32
num_workers=0

# prepare data loaders
train_loader = torch.utils.data.DataLoader(train_data, collate_fn=variable_collate, batch_size=batch_size, 
                                           num_workers=num_workers, shuffle=True)
val_loader   = torch.utils.data.DataLoader(val_data, collate_fn=variable_collate, batch_size=batch_size,
                                           num_workers=num_workers, shuffle=True)
test_loader  = torch.utils.data.DataLoader(test_data, collate_fn=variable_collate, batch_size=batch_size, 
                                           num_workers=num_workers, shuffle=True)

```

## Constructing the Model

I would like to use transfer learning with one of the big CNN's, but I need one that will use my chosen image size, or something close to it.
* VGG uses 224x224
* ResNet50 - not clear, but it looks like it can handle different sizes, up to 640 on short side.
* GoogLeNet 50 appears to take 224x224 also.

I'm starting to think all of these models are built for 224 square as a de facto standard for benchmarking.  Therefore, I choose to write my own model from scratch.  There is no doubt it will perform much worse than these world class models, but it will have one advantage in being able to use more input data (more pixels in the input images).

```python
import torch.nn as nn
import torch.nn.functional as F

class SkinCancerCnn(nn.Module):
    def __init__(self):
        super(SkinCancerCnn, self).__init__()
        
        # Applying batch norm to all but first and last layers, per guidance from the DL nanodegree
        # class, module 5.  Not clear why this restriction, however.
        
        #input images are 768x1024 pixels, full color (3 channels)
        image_height = 768
        image_width = 1024
        
        
        #########################
        # adjustable model params
        #########################

        conv1_out_channels = 8
        conv1_stride = 2
        self.conv1_pool = 1        # layer output 8 x 512 x 384
        
        conv2_out_channels = 16
        conv2_stride = 2
        self.conv2_pool = 1 #use 1 to indicate no pooling; layer output = 16 x 256 x 192
        
        conv3_out_channels = 32
        conv3_stride = 2
        self.conv3_pool = 1        # layer output 32 x 128 x 96
        
        conv4_out_channels = 64
        conv4_stride = 2
        self.conv4_pool = 1        # layer output 64 x 64 x 48
        
        conv5_out_channels = 128
        conv5_stride = 2
        self.conv5_pool = 1        # layer output 128 x 32 x 24
        
        channel_depth = conv5_out_channels #must be umber of out channels in final layer
        
        
        
        # compute the number of output features from the convolutional layers; wherever stride > 1
        # we need to reduce the size accordingly
        self.conv_out_features = int(channel_depth * image_height * image_width /
                                    (conv1_stride * conv1_stride * #square each for identical x & y dimensions
                                     conv2_stride * conv2_stride *
                                     conv3_stride * conv3_stride *
                                     conv4_stride * conv4_stride *
                                     conv5_stride * conv5_stride *
                                     self.conv1_pool * self.conv1_pool *
                                     self.conv2_pool * self.conv2_pool *
                                     self.conv3_pool * self.conv3_pool *
                                     self.conv4_pool * self.conv4_pool *
                                     self.conv5_pool * self.conv5_pool))
        print("conv_out_features = ", self.conv_out_features)
        
        self.c1 = nn.Conv2d(3, conv1_out_channels, 7, conv1_stride, padding=3)

        self.c2 = nn.Conv2d(conv1_out_channels, conv2_out_channels, 5, conv2_stride, padding=2, bias=False)
        self.b2 = nn.BatchNorm2d(conv2_out_channels)
        
        self.c3 = nn.Conv2d(conv2_out_channels, conv3_out_channels, 3, conv3_stride, padding=1, bias=False)
        self.b3 = nn.BatchNorm2d(conv3_out_channels)
        
        self.c4 = nn.Conv2d(conv3_out_channels, conv4_out_channels, 3, conv4_stride, padding=1, bias=False) #must not use bias if using batchnorm!
        self.b4 = nn.BatchNorm2d(conv4_out_channels)
        
        self.c5 = nn.Conv2d(conv4_out_channels, conv5_out_channels, 3, conv5_stride, padding=1, bias=False)
        self.b5 = nn.BatchNorm2d(conv5_out_channels)
        
        self.pool2 = nn.MaxPool2d(2, 2)
        self.pool3 = nn.MaxPool2d(3, 3)
        
        #linear classifier
        #self.fc1 = nn.Linear(self.conv_out_features, 3082)
        #self.fc2 = nn.Linear(3082, 96)
        #self.fcfinal = nn.Linear(96, 3)
        
        self.fc1 = nn.Linear(self.conv_out_features, 1024)
        self.fcfinal = nn.Linear(1024, 3)
        self.dropout = nn.Dropout(0.2)
        
        
    def forward(self, x):
        #memory_rpt("forward top")
        #print("      at entry, x.shape = ", x.shape)
        
        # conv layer 1
        x = F.leaky_relu(self.c1(x))
        if self.conv1_pool == 2:
            x = self.pool2(x)
        elif self.conv1_pool == 3:
            x = self.pool3(x)
        
        # Experimenting with placing the BN function before or after the relu in each layer.
        # It seems there is some debate about which is best.
        
        # conv layer 2
        x = self.c2(x)
        x = F.leaky_relu(x, negative_slope=0.05)
        x = self.b2(x)
        if self.conv2_pool == 2:
            x = self.pool2(x)
        elif self.conv2_pool == 3:
            x = self.pool3(x)
        
        # conv layer 3
        x = self.c3(x)
        x = F.leaky_relu(x, negative_slope=0.05)
        x = self.b3(x)
        if self.conv3_pool == 2:
            x = self.pool2(x)
        elif self.conv3_pool == 3:
            x = self.pool3(x)
        
        # conv layer 4
        x = self.c4(x)
        x = F.leaky_relu(x, negative_slope=0.05)
        x = self.b4(x)
        if self.conv4_pool == 2:
            x = self.pool2(x)
        elif self.conv4_pool == 3:
            x = self.pool3(x)
        
        # conv layer 5
        x = self.c5(x)
        x = F.leaky_relu(x, negative_slope=0.05)
        x = self.b5(x)
        if self.conv5_pool == 2:
            x = self.pool2(x)
        elif self.conv5_pool == 3:
            x = self.pool3(x)
        
        ### CONSIDER putting garbage collectiion in between these layers - worth some experimenting.
        #memory_rpt("  fwd conv ")
        
        conv_out = x.shape[1] * x.shape[2] * x.shape[3]
        #print("      after conv, x.shape = ", x.shape, "conv_out = ", conv_out)
        
        x = x.view(x.shape[0], -1) #flatten the convolved image
        #memory_rpt("  fwd view ")
        
        x = self.dropout(x)
        x = F.leaky_relu(self.fc1(x), negative_slope=0.05)
        
        #x = self.dropout(x)
        #x = F.leaky_relu(self.fc2(x), negative_slope=0.05)
        
        # softmax will be applied during training (as part of loss function), not here.
        x = self.dropout(x)
        x = self.fcfinal(x)
        
        return x
    
```

# Summary & Lessons Learned

The end result of this project was definitely not a useful descriminator for the 3 classes of data samples.  However, it taught me a lot about building a real-world ML model.  As a result, the process phases and durations are not likely to be terribly representative of future project that depend on good performance as a result.  In particular, some of the phases could be expected to take longer in order to achieve useful performance.  Some key lessons learned:

### Environment

1. Need to spend more time up front planning the resources to be used and getting them properly configured.  It is critical to understand how much space will be needed and allocate that before starting.  Count on model checkpoint files taking up a few hundred MB per copy.  Ubuntu 18 install, with other tools in an AMI can take > 70 GB by itself.

2. Keep large databases out of Github, or at least put them in a separate repo, so that the code can be easily pushed & pulled with minimal space & time constraints (esp for Udacity homework, where they have size limits on the whole repo).

3. Remember that Github limits files to 100 MB.  Therefore, checkpoints and image data may not fit into any repo.  Need another approach to moving data to/from AWS or other resources (e.g. S3?).

4. GPU info can be seen by calling torch.cuda.current_device() and torch.cuda.get_device_properties().  In AWS, p2.xlarge instance has one K80 GPU, which provides compute capability of 3.7, plus 61 GB memory for $0.9/hr.  What flexibility is there to choose GPU memory?

5. Remember to install .credentials and .gitconfig in order to access Github properly.

### Data

1. There are ways to make a NN handle variable input image sizes, but I elected not to do that extra work. The downsampling & cropping I did seems to have worked pretty well.  However, in such a study it might be better to manually (?) walk through each photo and crop around the area of interest, or to use something like yolo to identify the object and crop it.  There was a lot of variation in the size and placement of the objects of interest in the frames.  I'm sure dealing with that large variable amount of background made the training much more difficult.

2. Someone posted a suggestion that test data should be loaded with the same augmentations that training & validation data use.  I don't understand why this would make any difference, but my one experiment with it does support the idea.

### Model monitoring

1. Having the extra code in the training, validation & test sections to print aesthetic interim & final results is very valuable, including analysis of accuracy on each class, not just overall.

2. Saving the latest checkpoint after each epoch gives lots of peace of mind. Sometimes I wanted to kill a run and restart with a different LR, but the epoch of desire was not a multiple of 10, so didn't have a way to start from there.  Overall, the checkpoint saving mechanism in place is really good.

3. GPU memory use is really fickle, especially with Jupyter notebook (need to restart the Jupyter server if memory limit is reached).  Using the memory cleaning & monitoring code I put into my loops is critical.

### Model structure

1. Batchnorm2d adds a small amount of memory.  It is not obvious to me how much it affects model performance.  I did not see that it allowed a faster learning rate, as claimed, but maybe that was because my model was otherwise so far off the mark.

2. Batchnorm placement between layers (to normalize input to the next layer) seems to make most sense, but I didn't thoroughly test whether placing there is better than inside a layer (before its activation, as in the original paper).

3. Performance on GPS is not great.  I got maybe 25% faster than on my VM cpu.  I still wonder if I'm doing something wrong there.

4. Leaky RELU is more effective than simple ReLU.

5. Don't use any activation function on the final FC layer!  The optimizer will apply the softmax function to it.  

6. Be sure to move the model to gpu before declaring the optimizer (which uses model params).

7. Maxpool seems to make discrimination worse, rather than better.  Note that the world class models make little or no use of it.

8. Large FC layers add a LOT to the memory footprint (number of params)!  Use them sparingly.  It seemed that just one final FC layer was inadequate, but having two worked pretty well (didn't experiment much with this).

9. Conv layers should use kernels of odd dimensions.  It seems to be good to have different kernel sizes in different layers.  Lots of question remain on what are good sizes (11 or more in early layers with the large images I'm inputting?) and how the sizes should be sequenced, and how size & stride relate.
